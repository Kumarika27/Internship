```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
# Answer 4
```

```python
# Initialize the WebDriver
driver = webdriver.Chrome()
```

```python
# Navigate to the website
driver.get('https://www.flipkart.com/')
```

```python
# Search for Sunglasses
sunglass=driver.find_element(By.CLASS_NAME,'Pke_EE')
sunglass.send_keys('Sunglasses')
```

```python
# Click on the Search button
search=driver.find_element(By.CLASS_NAME,'_2iLD__')
search.click()
```

```python
brand=[]
protection=[]
description=[]
price=[]
```

```python
# Brand Name
brand_name=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
for i in brand_name[0:100]:
    brand_=i.text
    brand.append(brand_name)
# Protection
protect=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
for i in protect:
    protection_=i.text
    protection.append(protect)
#Price
rate=driver.find_elements(By.XPATH,'//div[@class="_3I9_wc"]')
for i in rate:
    rate_=i.text
    price.append(rate)
# Discount
discount=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
for i in discount:
    discount_=i.text
    description.append(discount)
```

```python
print(len(brand), len(protection), len(description), len(price))
```

```python
df=pd.DataFrame([brand, protection, description, price])
df.T
```
