```python
site3= requests.get('https://www.cnbc.com/world/?region=world')
site3
a=BeautifulSoup(site3.content)
a
headline= []
for i in a.find_all('a' ,class_="LatestNews-headline"):
    headline.append(i.text)
headline
time= []
for i in a.find_all('span' ,class_="LatestNews-wrapper"):
    time.append(i.text)
time
urls=[]
for i in a.find_all('a'):
    urls.append(i.get('href'))
urls
df=pd.DataFrame([headline, time,urls])
df.T
```
